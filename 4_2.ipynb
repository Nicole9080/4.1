```python
"""
Лабораторная работа 4.1
Вариант 15: Страховая компания
Выполнила Манидарова Н.Ч, ЦИБ-251
"""


class ValidationError(Exception):
    """Пользовательское исключение для ошибок валидации данных."""
    pass


class Policy:
    """
    Базовый класс для представления страхового полиса.
    Атрибуты:
        policy_id: Идентификатор полиса
        holder_name: Имя страхователя
        premium_amount: Сумма страховой премии
    """

    def __init__(self, policy_id, holder_name, premium_amount):
        """
        Инициализация объекта Policy (страховой премии).
        """
        # Валидация policy_id
        if not isinstance(policy_id, int):
            raise TypeError(f"policy_id должен быть целым числом, получено: {type(policy_id).__name__}")
        if policy_id <= 0:
            raise ValidationError(f"policy_id должен быть положительным числом, получено: {policy_id}")

        # Валидация holder_name
        if not isinstance(holder_name, str):
            raise TypeError(f"holder_name должен быть строкой, получено: {type(holder_name).__name__}")
        if not holder_name.strip():
            raise ValidationError("holder_name не может быть пустой строкой")

        # Валидация premium_amount
        if not isinstance(premium_amount, (int, float)):
            raise TypeError(f"premium_amount должен быть числом, получено: {type(premium_amount).__name__}")
        if premium_amount < 0:
            raise ValidationError(f"premium_amount не может быть отрицательным, получено: {premium_amount}")

        self.policy_id = policy_id
        self.holder_name = holder_name.strip()
        self.premium_amount = float(premium_amount)

    def calculate_payout(self, damage):
        """
        Полиморфный метод.
        Рассчитывает выплату по страховому случаю.
        В базовом классе возвращает 0 (будет переопределён в дочерних).
        """
        return 0

    def display_info(self):
        """Выводит информацию о полисе в консоль."""
        print(f"Полис ID: {self.policy_id}")
        print(f"Страхователь: {self.holder_name}")
        print(f"Сумма премии: {self.premium_amount} руб.")

    def __str__(self):
        return f"Полис #{self.policy_id} | {self.holder_name} | {self.premium_amount} руб."



class AutoInsurance(Policy):
    """
    Дочерний класс для автострахования.
    Дополнительные атрибуты:
        car_model: Модель автомобиля
        car_year: Год выпуска
    """

    def __init__(self, policy_id, holder_name, premium_amount, car_model, car_year):
        """
        Инициализация объекта AutoInsurance.
        """
        super().__init__(policy_id, holder_name, premium_amount)

        if not isinstance(car_model, str):
            raise TypeError(f"car_model должен быть строкой, получено: {type(car_model).__name__}")
        if not car_model.strip():
            raise ValidationError("car_model не может быть пустой строкой")

        if not isinstance(car_year, int):
            raise TypeError(f"car_year должен быть целым числом, получено: {type(car_year).__name__}")
        if car_year < 1900 or car_year > 2025:
            raise ValidationError(f"car_year должен быть между 1900 и 2025, получено: {car_year}")

        self.car_model = car_model.strip()
        self.car_year = car_year

    def calculate_payout(self, damage):
        """Рассчитывает выплату по автострахованию (80% от ущерба)."""
        return damage * 0.8

    def display_info(self):
        """Выводит информацию об автостраховании."""
        super().display_info()
        print(f"Модель авто: {self.car_model}")
        print(f"Год выпуска: {self.car_year}")

    def __str__(self):
        return f"Автострахование #{self.policy_id} | {self.holder_name} | {self.premium_amount} руб. | {self.car_model} ({self.car_year})"


class HealthInsurance(Policy):
    """
    Дочерний класс для медицинского страхования.
    Дополнительные атрибуты:
        coverage_type: Тип покрытия (например, "Стандарт", "Премиум")
        hospital_list: Список клиник
    """

    def __init__(self, policy_id, holder_name, premium_amount, coverage_type, hospital_list):
        """
        Инициализация объекта HealthInsurance.
        """
        super().__init__(policy_id, holder_name, premium_amount)

        if not isinstance(coverage_type, str):
            raise TypeError(f"coverage_type должен быть строкой, получено: {type(coverage_type).__name__}")
        if not coverage_type.strip():
            raise ValidationError("coverage_type не может быть пустой строкой")

        if not isinstance(hospital_list, list):
            raise TypeError(f"hospital_list должен быть списком, получено: {type(hospital_list).__name__}")
        if not hospital_list:
            raise ValidationError("hospital_list не может быть пустым")

        self.coverage_type = coverage_type.strip()
        self.hospital_list = hospital_list

    def calculate_payout(self, damage):
        """Рассчитывает выплату по медицинскому страхованию (90% от ущерба)."""
        return damage * 0.9

    def display_info(self):
        """Выводит информацию о медицинском страховании."""
        super().display_info()
        print(f"Тип покрытия: {self.coverage_type}")
        print(f"Клиники: {', '.join(self.hospital_list)}")

    def __str__(self):
        return f"Медицинское страхование #{self.policy_id} | {self.holder_name} | {self.premium_amount} руб. | {self.coverage_type}"



class Policyholder:
    """
    Базовый класс для представления клиента страховой компании.
    Атрибуты:
        client_id: Идентификатор клиента
        full_name: Полное имя клиента
        phone: Номер телефона
    """

    def __init__(self, client_id, full_name, phone):
        """
        Инициализация объекта Policyholder.
        """
        if not isinstance(client_id, int):
            raise TypeError(f"client_id должен быть целым числом, получено: {type(client_id).__name__}")
        if client_id <= 0:
            raise ValidationError(f"client_id должен быть положительным числом, получено: {client_id}")

        if not isinstance(full_name, str):
            raise TypeError(f"full_name должен быть строкой, получено: {type(full_name).__name__}")
        if not full_name.strip():
            raise ValidationError("full_name не может быть пустой строкой")

        if not isinstance(phone, str):
            raise TypeError(f"phone должен быть строкой, получено: {type(phone).__name__}")
        if not phone.strip():
            raise ValidationError("phone не может быть пустой строкой")

        self.client_id = client_id
        self.full_name = full_name.strip()
        self.phone = phone.strip()
        self.policies = []

    def add_policy(self, policy):
        """Добавляет полис клиенту."""
        if not isinstance(policy, Policy):
            raise TypeError(f"policy должен быть объектом Policy, получено: {type(policy).__name__}")
        self.policies.append(policy)

    def get_total_premium(self):
        """Возвращает общую сумму всех страховых премий клиента."""
        return sum(p.premium_amount for p in self.policies)

    def display_info(self):
        """Выводит информацию о клиенте в консоль."""
        print(f"Клиент: {self.full_name}")
        print(f"ID: {self.client_id}")
        print(f"Телефон: {self.phone}")
        print(f"Количество полисов: {len(self.policies)}")
        print(f"Общая сумма премий: {self.get_total_premium()} руб.")

    def __str__(self):
        return f"Клиент: {self.full_name} | Телефон: {self.phone} | Полисов: {len(self.policies)} | Общая премия: {self.get_total_premium()} руб."


class FamilyPlanHolder(Policyholder):
    """
    Дочерний класс для клиента с семейным планом страхования.
    Дополнительные атрибуты:
        family_members: Количество членов семьи
    """

    def __init__(self, client_id, full_name, phone, family_members):
        """
        Инициализация объекта FamilyPlanHolder.
        """
        super().__init__(client_id, full_name, phone)

        if not isinstance(family_members, int):
            raise TypeError(f"family_members должен быть целым числом, получено: {type(family_members).__name__}")
        if family_members < 1:
            raise ValidationError(f"family_members должен быть минимум 1, получено: {family_members}")

        self.family_members = family_members

    def get_family_discount(self):
        """
        Возвращает скидку в зависимости от количества членов семьи.
        """
        if self.family_members >= 4:
            return 0.15  # 15% скидка
        elif self.family_members >= 2:
            return 0.10  # 10% скидка
        else:
            return 0.0   # нет скидки

    def display_info(self):
        """Выводит информацию о клиенте с семейным планом."""
        super().display_info()
        print(f"Членов семьи: {self.family_members}")
        print(f"Семейная скидка: {self.get_family_discount() * 100}%")

    def __str__(self):
        return f"Семейный план: {self.full_name} | {self.family_members} чел. | Полисов: {len(self.policies)}"


class InsuranceCase:
    """
    Класс-контейнер для страхового случая.
    Хранит клиента и список заявленных полисов.
    """

    def __init__(self, policyholder):
        """Инициализация страхового случая. Принимает объект Policyholder."""
        if not isinstance(policyholder, Policyholder):
            raise TypeError(f"policyholder должен быть объектом Policyholder, получено: {type(policyholder).__name__}")

        self.policyholder = policyholder
        self.items = []

    def add_item(self, item):
        """Добавляет полис в страховой случай."""
        if not isinstance(item, Policy):
            raise TypeError(f"item должен быть объектом Policy, получено: {type(item).__name__}")
        self.items.append(item)

    def get_total_payout(self, damage):
        """
        Рассчитывает общую выплату по всем полисам с учётом семейной скидки.
        """
        total_payout = 0

        for item in self.items:
            payout = item.calculate_payout(damage)
            total_payout += payout
            print(f"  {item.__class__.__name__}: выплата = {payout:.2f} руб.")

        # Применяем скидку для FamilyPlanHolder
        if isinstance(self.policyholder, FamilyPlanHolder):
            discount = self.policyholder.get_family_discount()
            total_payout = total_payout * (1 - discount)
            print(f"  Семейная скидка: {discount * 100}%")

        return total_payout

    def display_report(self, damage):
        """Выводит детализированный отчёт по страховому случаю."""
        print("=" * 60)
        print("СТРАХОВОЙ СЛУЧАЙ - ДЕТАЛИЗИРОВАННЫЙ ОТЧЁТ")
        print("=" * 60)
        print()
        print(f"Клиент: {self.policyholder.full_name}")
        print(f"Телефон: {self.policyholder.phone}")
        if isinstance(self.policyholder, FamilyPlanHolder):
            print(f"Тип: Семейный план ({self.policyholder.family_members} чел.)")
        else:
            print("Тип: Обычный клиент")
        print()

        print(f"Размер ущерба: {damage:.2f} руб.")
        print()
        print("Заявленные полисы:")
        print("-" * 40)
        for item in self.items:
            print(f"  {item}")
        print()

        print("Расчёт выплат:")
        print("-" * 40)
        total_payout = self.get_total_payout(damage)
        print()
        print(f"ИТОГОВАЯ ВЫПЛАТА: {total_payout:.2f} руб.")
        print("=" * 60)


# ============================================================
# ОСНОВНАЯ ФУНКЦИЯ (Демонстрация)
# ============================================================

def main():
    """Основная функция для демонстрации работы классов."""

    print("=" * 60)
    print("СОЗДАНИЕ ПОЛИСОВ")
    print("=" * 60)
    print()

    # Создание полисов
    try:
        auto1 = AutoInsurance(1, "Мелина Екатерина", 15000.00, "Toyota Camry", 2022)
        print(f"Создан: {auto1}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        auto1 = None

    try:
        auto2 = AutoInsurance(2, "Любова Рената", 12000.00, "Hyundai Solaris", 2021)
        print(f"Создан: {auto2}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        auto2 = None

    try:
        health1 = HealthInsurance(3, "Реджистер Юрий", 25000.00, "Премиум", ["Клиника №1", "Клиника №2"])
        print(f"Создан: {health1}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        health1 = None

    try:
        health2 = HealthInsurance(4, "Мелина Екатерина", 18000.00, "Стандарт", ["Клиника №1"])
        print(f"Создан: {health2}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        health2 = None

    print()
    print("=" * 60)
    print("СОЗДАНИЕ КЛИЕНТОВ")
    print("=" * 60)
    print()

    # Создание клиентов
    try:
        client1 = Policyholder(1, "Мелина Екатерина Юрьевна", "+7-916-102-58-23")
        print(f"Создан: {client1}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        client1 = None

    try:
        client2 = FamilyPlanHolder(2, "Любова Рената Ахмедовна", "+7-916-345-27-88", 3)
        print(f"Создан: {client2}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        client2 = None

    try:
        client3 = FamilyPlanHolder(3, "Реджистер Юрий Дмитриевич", "+7-999-987-65-43", 5)
        print(f"Создан: {client3}")
    except (TypeError, ValidationError) as e:
        print(f"Ошибка: {e}")
        client3 = None

    print()
    print("=" * 60)
    print("ПРИВЯЗКА ПОЛИСОВ К КЛИЕНТАМ")
    print("=" * 60)
    print()

    # Привязка полисов к клиентам
    if client1 is not None:
        if auto1 is not None:
            client1.add_policy(auto1)
            print(f"Клиенту {client1.full_name} добавлен полис {auto1.policy_id}")
        if health2 is not None:
            client1.add_policy(health2)
            print(f"Клиенту {client1.full_name} добавлен полис {health2.policy_id}")

    if client2 is not None and auto2 is not None:
        client2.add_policy(auto2)
        print(f"Клиенту {client2.full_name} добавлен полис {auto2.policy_id}")

    if client3 is not None and health1 is not None:
        client3.add_policy(health1)
        print(f"Клиенту {client3.full_name} добавлен полис {health1.policy_id}")

    print()
    print("=" * 60)
    print("ДЕМОНСТРАЦИЯ ПОЛИМОРФИЗМА (calculate_payout)")
    print("=" * 60)
    print()

    damage_amount = 100000

    print(f"Размер ущерба: {damage_amount} руб.")
    print()
    print("Расчёт выплат по полисам:")
    print("-" * 40)

    if auto1 is not None:
        print(f"  AutoInsurance: выплата = {auto1.calculate_payout(damage_amount):.2f} руб.")
    if auto2 is not None:
        print(f"  AutoInsurance: выплата = {auto2.calculate_payout(damage_amount):.2f} руб.")
    if health1 is not None:
        print(f"  HealthInsurance: выплата = {health1.calculate_payout(damage_amount):.2f} руб.")
    if health2 is not None:
        print(f"  HealthInsurance: выплата = {health2.calculate_payout(damage_amount):.2f} руб.")

    print()
    print("=" * 60)
    print("ДЕМОНСТРАЦИЯ КЛАССА-КОНТЕЙНЕРА (InsuranceCase)")
    print("=" * 60)
    print()

    # Страховой случай для клиента 3 (семейный план, 5 человек)
    if client3 is not None and health1 is not None:
        insurance_case = InsuranceCase(client3)
        insurance_case.add_item(health1)
        insurance_case.display_report(damage_amount)

    print()

    # Страховой случай для клиента 2 (семейный план, 3 человека)
    if client2 is not None and auto2 is not None:
        insurance_case2 = InsuranceCase(client2)
        insurance_case2.add_item(auto2)
        insurance_case2.display_report(damage_amount)


if __name__ == "__main__":
    main()
```
